**SELECTED FEATURES USING RFE**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import numpy as np

Y = np.ravel(y)
X_train, X_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

classifier = LogisticRegression()

rfe = RFE(estimator=classifier, n_features_to_select=20)

# Fit the RFE selector to the training data
rfe.fit(X_train, Y_train)

# Get the selected features
selected_features = X.columns[rfe.support_]

# Print the selected features
print("Selected features:", selected_features)
```

**PRINCIPAL COMPONENT ANALYSIS**

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)


X_pca = pca.fit_transform(X_scaled)

X_pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])

print("Explained variance ratio:", pca.explained_variance_ratio_)

print("Cumulative explained variance ratio:", pca.explained_variance_ratio_.cumsum())

print("Transformed features (principal components):\n", X_pca_df.head())
```

**HANDLING OUTLIERS USING Z SCORE**

```python
import pandas as pd
import numpy as np

z_scores = (X_scaled_df - X_scaled_df.mean()) / X.std()

z_score_threshold = 7

outliers = (z_scores > z_score_threshold) | (z_scores < -z_score_threshold)

outlier_counts = outliers.sum(axis=1)

X_no_outliers = X[outlier_counts == 0]

# the shape of the original and cleaned datasets
print("Original dataset shape:", X.shape)
print("Cleaned dataset shape:", X_no_outliers.shape)
```

**K FOLD**

```python

from sklearn.model_selection import KFold


X = np.array(X)
y = np.array(y)

if y.ndim > 1 and y.shape[1] == 1:
    y = y.ravel()

k_folds = 9
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

accuracy_scores = []

for fold, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"Fold {fold + 1}:")

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    classifier= CatBoostClassifier ()
    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")
    accuracy_scores.append(accuracy)

```

```python

# Calculate and print the mean accuracy across all folds
mean_accuracy = np.mean(accuracy_scores)
print(f"\nMean Accuracy Across {k_folds}-fold Cross-Validation: {mean_accuracy:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(range(1, k_folds + 1), accuracy_scores, color='blue')
plt.axhline(y=mean_accuracy, color='red', linestyle='--', label=f'Mean Accuracy: {mean_accuracy:.4f}')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Accuracy Scores for each Fold')
plt.legend()
plt.show()
```
